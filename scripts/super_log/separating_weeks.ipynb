## Important Functions

```python
def get_number_cat(cat):
    output = 0
    for i in list(df_super.columns):
        if cat in i:
            output += 1
    return int(output/2)
```

```python
def get_cols_week_no_ts(week, cols=list(df.columns)):
    output = []
    for i in cols:
        if 'week_' + str(week) in i and 'ts' not in i:
            output.append(i)
    #output.append('course_passing_state')
    #output.append('course_passed_items')
    #output.append('course_grade')
    return output
    
```

```python
# From: Generico/main.ipynb
def get_columns(columns, string):
    output = []
    for col in list(columns):
        if string in col:
            output.append(col)
    return output
```

```python
def get_completed(row, category):
    output = 0
    for i in row[category]:
        if i == 2:
            output += 1
    return output
```

```python
def get_grade(row):
    try:
        return np.mean(row)
    except:
        return 0
```

```python
def get_min(row, ts):
    try:
        return min(list(filter(lambda v: v==v, row[ts])))
    except:
        return 0
```

```python
def get_max(row, ts):
    try:
        return max(list(filter(lambda v: v==v, row[ts])))
    except:
        return 0
```

## Main

```python
# Course Metadata
metadata = {'course_name': course_name, 
            'course_total_weeks': number_weeks, 
            'course_total_lectures': int(len(get_columns(df.columns, 'lecture'))/2), 
            'course_total_supplements': int(len(get_columns(df.columns, 'supplement'))/2), 
            'course_total_quiz': int(len(get_columns(df.columns, 'quiz'))/2), 
            'course_total_exams': int(len(get_columns(df.columns, 'exam'))/2), 
            'course_total_phased_peer': int(len(get_columns(df.columns, 'phased_peer'))/2)}
```

```python
# Adding Course Charactetistics
df['course_name'] = metadata['course_name']
df['course_total_weeks'] = metadata['course_total_weeks']

df['course_total_lectures'] = metadata['course_total_lectures']
df['course_total_supplements'] = metadata['course_total_supplements']
df['course_total_quiz'] = metadata['course_total_quiz']
df['course_total_exams'] = metadata['course_total_exams']
df['course_total_phased_peer'] = metadata['course_total_phased_peer']

df['course_passable_items'] = metadata['course_total_exams'] + metadata['course_total_phased_peer']
```

```python
# Rename Columns
df.rename(columns={'course_passing_state_id': 'course_passing_state', 
                   'course_grade_verified_passed_items': 'course_passed_items', 
                   'course_grade_verified': 'course_grade'}, 
         inplace=True)
```

```python
# Adding percentage_passed_items
df['course_percentage_passed_items'] = df['course_passed_items'] / df['course_passable_items']
```

```python
cols_actual = get_cols_week_no_ts(1)
supplements = get_columns(cols_actual, 'supplement')
len(supplements)
```

```python
df_lectures = pd.DataFrame()
df_supplements = pd.DataFrame()
df_exams = pd.DataFrame()
df_quiz = pd.DataFrame()
df_phased_peer = pd.DataFrame()

for actual_week in range(1, metadata['course_total_weeks'] + 1):
    # Columns of the actual week
    cols_actual = get_cols_week_no_ts(actual_week)
    
    ### LECTURES INFO ###
    lectures = get_columns(cols_actual, 'lecture')
    df_lectures_aux = pd.DataFrame(index=df.index)
    
    for index, row in df[lectures].fillna(0).iterrows():
        df_lectures_aux.at[index, 'week_lectures_completed'] = list(row.values).count(2)
        df_lectures_aux.at[index, 'week_lectures_started'] = list(row.values).count(1)
        df_lectures_aux.at[index, 'week_lectures_no_interacted'] = list(row.values).count(0)
    df_lectures_aux['week_total_lectures'] = len(lectures)
    
    ### SUPPLEMENTS INFO ###
    supplements = get_columns(cols_actual, 'supplement')
    df_supplements_aux = pd.DataFrame(index=df.index)
    
    for index, row in df[supplements].fillna(0).iterrows():
        df_supplements_aux.at[index, 'week_supplements_completed'] = list(row.values).count(2)
        df_supplements_aux.at[index, 'week_supplements_started'] = list(row.values).count(1)
        df_supplements_aux.at[index, 'week_supplements_no_interacted'] = list(row.values).count(0)
    df_supplements_aux['week_total_supplements'] = len(supplements)
    
            
    ### EXAMS INFO ###
    exams = get_columns(cols_actual, 'exam')
    df_exams_aux = pd.DataFrame(index=df.index)
    for index, row in df[exams].fillna('NOT').iterrows():
        df_exams_aux.at[index, 'week_exams_no_interacted'] = list(row.values).count('NOT')
        interacted = [x for x in row.values if x != 'NOT']
        
        if len(interacted):
            df_exams_aux.at[index, 'week_exams_interacted'] = len(interacted)
            df_exams_aux.at[index, 'week_exams_grade'] = get_grade(interacted)
        
        else: # No interacted
            df_exams_aux.at[index, 'week_exams_interacted'] = 0
            df_exams_aux.at[index, 'week_exams_grade'] = 0
    df_exams_aux['week_total_exams'] = len(exams)
    
    ### QUIZ INFO ###
    quiz = get_columns(cols_actual, 'quiz')
    df_quiz_aux = pd.DataFrame(index=df.index)
    for index, row in df[quiz].fillna('NOT').iterrows():
        df_quiz_aux.at[index, 'week_quiz_no_interacted'] = list(row.values).count('NOT')
        interacted = [x for x in row.values if x != 'NOT']
        
        if len(interacted):
            df_quiz_aux.at[index, 'week_quiz_interacted'] = len(interacted)
            df_quiz_aux.at[index, 'week_quiz_grade'] = get_grade(interacted)
        
        else: # No interacted
            df_quiz_aux.at[index, 'week_quiz_interacted'] = 0
            df_quiz_aux.at[index, 'week_quiz_grade'] = 0   
    df_quiz_aux['week_total_quiz'] = len(quiz)
    
    ### PHASED PEER INFO ###
    phased_peer = get_columns(cols_actual, 'phased_peer')
    df_phased_peer_aux = pd.DataFrame(index=df.index)
    for index, row in df[phased_peer].fillna('NOT').iterrows():
        df_phased_peer_aux.at[index, 'week_phased_peer_no_interacted'] = list(row.values).count('NOT')
        interacted = [x for x in row.values if x != 'NOT']
        
        if len(interacted):
            df_phased_peer_aux.at[index, 'week_phased_peer_interacted'] = len(interacted)
            df_phased_peer_aux.at[index, 'week_phased_peer_grade'] = get_grade(interacted)
        
        else: # No interacted
            df_phased_peer_aux.at[index, 'week_phased_peer_interacted'] = 0
            df_phased_peer_aux.at[index, 'week_phased_peer_grade'] = 0   
    df_phased_peer_aux['week_phased_peer_quiz'] = len(phased_peer)

    
    # Actual week!
    df_lectures_aux['week_actual'] = actual_week
    df_supplements_aux['week_actual_s'] = actual_week
    df_exams_aux['week_actual_e'] = actual_week
    df_quiz_aux['week_actual_q'] = actual_week
    df_phased_peer_aux['week_actual_p'] = actual_week

    
    # Concatinig different weeks
    df_lectures = pd.concat([df_lectures, df_lectures_aux])
    df_supplements = pd.concat([df_supplements, df_supplements_aux])
    df_exams = pd.concat([df_exams, df_exams_aux])  
    df_quiz = pd.concat([df_quiz, df_quiz_aux])  
    df_phased_peer = pd.concat([df_phased_peer, df_phased_peer_aux])    
    
    
    # Free memory
    del df_lectures_aux
    del df_supplements_aux
    del df_exams_aux
    del df_quiz_aux
    del df_phased_peer_aux
    
    
```

```python
# Lectures
df_lectures['week_lectures_interacted'] = df_lectures['week_lectures_completed'] + df_lectures['week_lectures_started']

df_lectures['week_percentage_lectures_completed'] = df_lectures['week_lectures_completed'] / df_lectures['week_total_lectures']
df_lectures['week_percentage_lectures_started'] = df_lectures['week_lectures_started'] / df_lectures['week_total_lectures']
df_lectures['week_percentage_lectures_no_interacted'] = df_lectures['week_lectures_no_interacted'] / df_lectures['week_total_lectures']
df_lectures['week_percentage_lectures_interacted'] = df_lectures['week_lectures_interacted'] / df_lectures['week_total_lectures']

df_lectures.fillna(0, inplace=True)
```
